```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jun 13 17:58:35 2020

@author: Owner
"""


import urllib.request
from bs4 import BeautifulSoup
#Enter the name of the song,movie name or singer name or movie name
String=input('Enter the string')
search=String.replace(" ", "+")
string='https://www.youtube.com/results?search_query='+search
URL=string
soup=BeautifulSoup(urllib.request.urlopen(URL).read())
soup.prettify()
#To store the values 
title=[]
Duration=[]
Year=[]
Views=[]
#Fetch the values from the div tag having the below class
results = soup.find_all('div',class_='yt-lockup-content')
for i in results:
    #Fetch the values from the a tag in the div tag
    links=i.find_all('a',class_='yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink spf-link')
    #Fetch the values from the span tag in the div tag
    duration=i.find_all('span',class_="accessible-description")
    #Fetch the values from the div tag in the div tag
    data=i.find_all('div',class_='yt-lockup-meta')
    #To maintain counters whether the tags have vaules or not
    counter='0'
    titleno='0'
    viewsno='0'
    yearno='0'
    durationno='0'
    
    for j in links:
        titleno='1'
        title.append(j.get("title"))
        
    for k in duration:
        durationno='1'
        lengthDuration=len(k.string)
        dur=k.string[13:lengthDuration]
        Duration.append(dur)

    for n in data:
        counter='1'
        data1=n.find_all('li')
        for p in data1:
            length=len(p.text)
            last=length-5
            lastdays=length-3
            p.text[last:length]
            

            if(p.text[lastdays:length]=='ago'):
                yearno='1'
                year=p.text[0:last]
                Year.append(year)
            if(p.text[last:length]=='views'):
                viewsno='1'
                views=p.text[0:last]
                Views.append(views)
                
    if(counter=='0'):
        Views.append(" ")
        Year.append(" ")
    if(titleno=='0'):
        title.append("No title")
    if(viewsno=='0' and counter=='1'):
        Views.append(" ")
    if(yearno=='0' and counter=='1'):
        Year.append(" ")
    if(durationno=='0'):
        Duration.append(" ")

import pandas as pd
Youtube = pd.DataFrame({'Title': title,'Duration':Duration,
                              'Views': Views,'Year':Year})
```
